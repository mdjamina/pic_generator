```python
import wordcloud as wc
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
import matplotlib.pyplot as plt
import nltk
nltk.download('punkt')
nltk.download('stopwords')
import pprint as pp

from nltk import word_tokenize
from nltk.corpus import stopwords
import string 


```

```python
french_stopwords = set(stopwords.words('french'))
```

```python
def tokenizer(text,language = "french"):
    """onvertir text en une liste de mots/tokens     
    """
    lst = word_tokenize(text, language = language)
    return [w.lower() for w in lst if w not in string.punctuation]
```

```python
def pic_generator(text,pic_file_name='wordcloudimage.png', max_words=250,max_font_size=250,stopwords=None,background_color='white',width=1000,height=720):

  wordcloudimage = WordCloud( max_words=max_words,font_step=2 , max_font_size=max_font_size, stopwords=stopwords, background_color=background_color, width=width, height=height, margin=0 ).generate(text)

  plt.figure(figsize=(20,8))
  plt.imshow(wordcloudimage)
  plt.axis("off")
  plt.show()
  wordcloudimage.to_file(pic_file_name)

          
```
